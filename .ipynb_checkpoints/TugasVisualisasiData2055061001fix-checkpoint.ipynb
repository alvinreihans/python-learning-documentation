```python
# Importing the matplotlib library
import numpy as np
import matplotlib.pyplot as plt
# Declaring the figure or the plot (y, x) or (width, height)
plt.figure(figsize=[15, 10])
# Data to be plotted
twitter = [54, 117, 185, 241, 288, 305, 318, 330, 321, 340, 353]
facebook = [608, 845, 1056, 1228, 1393, 1591, 1860, 2129, 2320, 2498, 2797]
instagram = [0, 0, 0, 150, 300, 400, 510.9, 593.7, 766.4, 932.7, 1221]
# Using numpy to group 3 different data with bars
X = np.arange(len(twitter))
# Passing the parameters to the bar function, this is the main function which creates the bar plot
# Using X now to align the bars side by side
plt.bar(X + 0.25*X, twitter, color = 'g', width = 0.25)
plt.bar(X + 0.25*X + 0.25, facebook, color = 'b', width = 0.25)
plt.bar(X + 0.25*X + 0.50, instagram, color = 'm', width = 0.25)
# Creating the legend of the bars in the plot
plt.legend(['Twitter', 'Facebook', 'Instagram'])
# Overiding the x axis with year
plt.xticks([i + 0.25*i + 0.25 for i in range(11)], [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])
# Giving the tilte for the plot
plt.title("Bar plot representing Twitter, Facebook and Instagram")
# Namimg the x and y axis
plt.xlabel('Year')
plt.ylabel('User')
# Saving the plot as a 'png'
plt.savefig('4BarPlot.png')
# Displaying the bar plot
plt.show()


# In[ ]:
```
